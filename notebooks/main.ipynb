```python
relations_store = rendu.aggregate_relations_from_jsonl(
    DATA_DIR / "data_arcane.json",
    rendu.extract_relations
)

print(f"Total unique relations: {len(relations_store)}")

for (h, rel, t), data in list(relations_store.items()):
    print(f"{h} --[{rel}]--> {t}")
```

```python
from collections import defaultdict

relations_count = {}
# Use a defaultdict of sets to easily group unique 't' (tails) for each 'h' (head)
relations_links = defaultdict(set)

# 1. Group all target entities (t) by their head entity (h)
for (h, rel, t), data in relations_store.items():
    relations_links[h].add(t)
    relations_links[t].add(h)

# 2. Count the number of links for each head entity
for h, tails in relations_links.items():
    relations_count[h] = len(tails)

# Optional: Print the results to verify
print(relations_count)
```

```python
def get_triplet_count(entity_str):
    """Calculates how many triplets an entity is involved in within kg_graph."""
    uri = rendu.ARCANE[rendu.clean_uri_string(entity_str)]
    as_subject = len(list(kg_graph.triples((uri, None, None))))
    as_object = len(list(kg_graph.triples((None, None, uri))))
    return as_subject + as_object
```

```python
sorted_entities_list = sorted(
    entities_store.items(), 
    key=lambda item: get_triplet_count(item[0]), 
    reverse=True
)

TOP_N = 15
important_entities = dict(sorted_entities_list[:TOP_N])

# Optional: Print them out to verify it worked and see your central nodes!
print(f"--- Top {TOP_N} Most Connected Entities ---")
for ent in important_entities:
    print(f"- {ent}: {get_triplet_count(ent)} triplets")    
```

```python
expanded_graph = openKB.expand_knowledge_graph(kg_graph, important_entities, max_entities=5)

print(f"\nExpanded Graph now has {len(expanded_graph)} total triplets.")

# Save the newly expanded graph to a new file
expanded_rdf_output_path = DATA_DIR / "arcane_knowledge_graph_expanded.ttl"
rendu.save_rdf_graph(expanded_graph, expanded_rdf_output_path)
print(f"Expanded RDF graph saved to {expanded_rdf_output_path}")
```

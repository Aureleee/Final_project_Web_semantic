```python
from pathlib import Path

DATA_DIR = Path.cwd() / ".." / "data"
arcane_engine = nlp.ArcaneNLP(use_gpu=False)
```

```python
entities_store, relations_store = nlp.extract_all_pages(
    start_urls=nlp.urls_arcane, 
    dir=DATA_DIR, 
    arcane_engine=arcane_engine,
    max_pages=50, # <-- This is the new parameter! Increase it if you want more data.
    output_file=DATA_DIR / "data.jsonl"
)
```

```python
nlp.save_entities_store_to_csv(entities_store, filename=DATA_DIR / "extracted_knowledge.csv")
print(f"Total unique relations: {len(relations_store)}")

for (h, rel, t), data in list(relations_store.items()):
    print(f"{h} --[{rel}]--> {t}")
```

```python
from collections import defaultdict

relations_count = {}
# Use a defaultdict of sets to easily group unique 't' (tails) for each 'h' (head)
relations_links = defaultdict(set)

# 1. Group all target entities (t) by their head entity (h)
for (h, rel, t), data in relations_store.items():
    relations_links[h].add(t)
    relations_links[t].add(h)
  
# 2. Count the number of links for each head entity
for h, tails in relations_links.items():
    relations_count[h] = len(tails)

# Optional: Print the results to verify
print(relations_count)
```

```python
print("---- RDF graph ----")
kg_graph = rdf.build_rdf_graph(entities_store, relations_store)

print(f"Graph built with {len(kg_graph)} triplets.")

kg_graph = rdf.clean_noisy_entities(kg_graph)
```

```python
rdf_output_path = DATA_DIR / "arcane_knowledge_graph.ttl"

nlp.save_rdf_graph(kg_graph, rdf_output_path)
print(f"RDF graph saved to {rdf_output_path}")

print("\nSample Triplets:")
for i, (s, p, o) in enumerate(kg_graph):
    if i >= 5: break
    print(f"{s} --[{p}]--> {o}")
```

```python
kg_graph = openKB.enrich_class_definitions(kg_graph, nlp.ARCANE)

# Save the perfectly aligned graph
output_path = DATA_DIR / "arcane_knowledge_graph_aligned.ttl"
nlp.save_rdf_graph(kg_graph, output_path)
print(f"\nSaved aligned graph to {output_path}")
```

```python
kg_path = DATA_DIR / "arcane_knowledge_graph_aligned.ttl"

# Folder where we will save the TSV splits
kge_data_dir = DATA_DIR / "kge_data"

print("1. Preparing Datasets for Machine Learning...")
train, valid, test = kge.prepare_kge_datasets(output_path, kge_data_dir)

print("\nSample of training data (head, relation, tail):")
print(train.head())
```

```python
kge_dir = DATA_DIR / "kge_data"
train_tsv = str(kge_dir / "train.tsv")
valid_tsv = str(kge_dir / "valid.tsv")
test_tsv = str(kge_dir / "test.tsv")

results_metrics = {}

# TransE 
transe_res, transe_metrics = kge.run_kge_experiment(
    train_tsv, valid_tsv, test_tsv, model_name='TransE', epochs=100
)
results_metrics['TransE'] = transe_metrics

# ComplEx 
complex_res, complex_metrics = kge.run_kge_experiment(
    train_tsv, valid_tsv, test_tsv, model_name='ComplEx', epochs=100
)
results_metrics['ComplEx'] = complex_metrics

# Final Comparison Table
comparison_df = kge.generate_comparison_report(results_metrics)
print("\n--- FINAL EVALUATION RESULTS ---")
display(comparison_df)
```
